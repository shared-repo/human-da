```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras as tf_keras
```

```python
# 입력 자동화 도구 만들기

# 특정 디렉터리의 파일을 읽어서 모델의 입력데이터로 변환하는 도구
train_generator = tf_keras.preprocessing.image.ImageDataGenerator(rescale=1/255, validation_split=0.2)
test_generator = tf_keras.preprocessing.image.ImageDataGenerator(rescale=1/255)
```

```python
# 입력 데이터 구성

datasets = []

dataset = train_generator.flow_from_directory(directory=f'sample_data/iic/seg_train/seg_train',
                                              target_size=(150, 150),
                                              batch_size=32,
                                              class_mode="categorical",
                                              subset="training")
datasets.append(dataset)

dataset = train_generator.flow_from_directory(directory=f'sample_data/iic/seg_train/seg_train',
                                              target_size=(150, 150),
                                              batch_size=32,
                                              class_mode="categorical",
                                              subset="validation")
datasets.append(dataset)

dataset = test_generator.flow_from_directory(directory=f'sample_data/iic/seg_test/seg_test',
                                             target_size=(150, 150),
                                             batch_size=32,
                                             class_mode="categorical")
datasets.append(dataset)
```
