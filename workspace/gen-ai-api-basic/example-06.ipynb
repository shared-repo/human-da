```python
import pandas as pd

data = [
    "이제 집에 가고 싶어요",
    "배가 고파서 저녁을 빨리 먹고 싶어요",
    "이러다가 굶어서 죽을 것 같아요",
    "퇴근시간이 가까워지고 있습니다",
    "트랜스포머는 시퀀스-시퀀스 구현 모델입니다",
    "저녁으로는 어니언치킨이 땡기네요"
]
df = pd.DataFrame(data, columns=['text'])
df
```

```python
# df['embedding'] = df.apply(lambda row: embeddings.embed_query(row["text"]), axis=1)
df['embedding'] = df['text'].map(lambda v: embeddings.embed_query(v))
df
```

```python
import numpy as np
def cos_sim(A, B):
  return np.dot(A, B)/(np.linalg.norm(A)*np.linalg.norm(B))

# query = "맛있는 음식을 먹는 것은 언제나 즐겁습니다."
query = "자연어 처리는 떠오르는 딥러닝 주제입니다."

embedded_query = embeddings.embed_query(query)
df['similarity'] = df['embedding'].map(lambda v: cos_sim(v, embedded_query))

df.sort_values(by="similarity", ascending=False)
```
