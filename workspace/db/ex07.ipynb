```python
import pymysql
```

```python
# 1. 데이터베이스에 연결
conn = pymysql.connect(host="127.0.0.1", port=3306, db="bookstore",
                       user="humanda", passwd="humanda")
```

```python
# 2. 명령 도구 준비
cursor = conn.cursor()
```

```python
# 3. 명령 실행
sql = "select * from customer"
cursor.execute(sql)
```

```python
# 4. 조회 명령 실행 결과를 가져오기
data = cursor.fetchall()
```

```python
# 5. 데이터 사용
# print( data )
# print( type(data), len(data) )
for row in data:
    print(row)
```

```python
# 6. 연결 닫기
cursor.close()
conn.close()
```

```python
# 1. 데이터베이스에 연결
conn = pymysql.connect(host="127.0.0.1", port=3306, db="employees",
                       user="root", passwd="0000")
```

```python
# 2. 명령 도구 준비
cursor = conn.cursor()
```

```python
# 3. 명령 실행
sql = "select emp_no, max(salary) from salaries group by emp_no"
cursor.execute(sql)
```
